## Dados e Gráficos dos experimentos

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Lê os dados obtidos pelos teste nos arquivos *.txt* e extrai os tempos de execução



```python
nCons = 34
nExec = 30

# Mude para o arquivo contendo os dados brutos do MySQL e do PostgreSQL desejado
mysql_file = "/content/mysql_no_index.txt"

psql_file = "/content/psql_no_index.txt"


with open(mysql_file, "r") as dataMySQL:
  timesMySQL = np.zeros((nCons, nExec))

  row = 0
  col = 0
  dataMySQL.readline()
  for line in dataMySQL:

    timesMySQL[row][col] = line.split("|")[-1].strip()

    row += 1
    if(row % nCons == 0):
      row = 0
      col += 1


with open(psql_file, "r") as dataPostgreSQL:
  timesPostgreSQL = np.zeros((nCons, nExec))

  row = 0
  col = 0
  dataPostgreSQL.readline()
  for line in dataPostgreSQL:

    timesPostgreSQL[row][col] = line.split("|")[-1].strip()

    row += 1
    if(row % nCons == 0):
      row = 0
      col += 1
```

## Mostra os tempos de execução do MySQL para cada uma das 30 execuções

```python
print("MySQL\n\n")

timesMySQL = np.concatenate([timesMySQL[:18], timesMySQL[21:]])
timesMySQL = timesMySQL[:-9]
i = 1
for x in timesMySQL:
  print("Consulta", i, end=':  ')
  if(i < 10): print(end=' ')
  for y in x:
    print(y, end=' ')
  print(end='\n\n')
  i += 1
```

## Mostra os tempos de execução do PostgreSQL para cada uma das 30 execuções

```python
print("PostgreSQL\n\n")

timesPostgreSQL = np.concatenate([timesPostgreSQL[:18], timesPostgreSQL[21:]])
timesPostgreSQL = timesPostgreSQL[:-9]
i = 1
for x in timesPostgreSQL:
  print("Consulta", i, end=':  ')
  if(i < 10): print(end=' ')
  for y in x:
    print(y, end=' ')
  print(end='\n\n')
  i += 1
```

## Boxplot dos tempos de execução do MySQL

```python
fig = plt.figure(figsize =(13, 3.9), dpi=150)

ax = fig.add_axes([0, 0, 0.5, 1])
ax.yaxis.grid(True, linestyle="--")
ax.set_title("MySQL")

aux = timesMySQL[:].transpose()

bp = ax.boxplot(aux, sym='.')

plt.xlabel("Consulta")
plt.ylabel("Tempo de Execução (s)")
plt.show()
```

## Boxplot dos tempos de execução do PostgreSQL

```python
fig = plt.figure(figsize =(13, 3.9), dpi=150)

ax = fig.add_axes([0, 0, 0.5, 1])
ax.yaxis.grid(True, linestyle="--")
ax.set_title("PostgreSQL")

aux = timesPostgreSQL[:].transpose()

bp = ax.boxplot(aux, sym='.')

plt.xlabel("Consulta")
plt.ylabel("Tempo de Execução (s)")
plt.show()
```

## Calcula tempo médio e desvio padrão para as consultas do MySQL

```python
print("MySQL\n\n")

mediasMySQL = []
dpsMySQL = []
varMySQL = []
sumMediasMySQL = 0

i = 1
for consulta in timesMySQL:
  print("Tempo Médio Consulta", i, end=':    ')
  if(i < 10): print(end=' ')
  print(np.mean(consulta), end='s\n')
  mediasMySQL.append(np.mean(consulta))
  sumMediasMySQL += np.mean(consulta)

  print("Desvio Padrão Consulta", i, end=':  ')
  if(i < 10): print(end=' ')
  print(np.std(consulta), end='s\n')
  dpsMySQL.append(np.std(consulta))

  print("Variância Consulta", i, end=':      ')
  if(i < 10): print(end=' ')
  print(np.var(consulta), end='s\n\n')
  varMySQL.append(np.var(consulta))

  i += 1

print("Média dos Tempos Médios de Consulta:", sumMediasMySQL/nCons)
```

## Calcula tempo médio e desvio padrão para as consultas do PostgreSQL

```python
print("PostgreSQL\n\n")

mediasPostgreSQL = []
dpsPostgreSQL = []
varPostgreSQL = []
sumMediasPostgreSQL = 0

i = 1
for consulta in timesPostgreSQL:
  print("Tempo Médio Consulta", i, end=':    ')
  if(i < 10): print(end=' ')
  print(np.mean(consulta), end='s\n')
  mediasPostgreSQL.append(np.mean(consulta))
  sumMediasPostgreSQL += np.mean(consulta)

  print("Desvio Padrão Consulta", i, end=':  ')
  if(i < 10): print(end=' ')
  print(np.std(consulta), end='s\n')
  dpsPostgreSQL.append(np.std(consulta))

  print("Variância Consulta", i, end=':      ')
  if(i < 10): print(end=' ')
  print(np.var(consulta), end='s\n\n')
  varPostgreSQL.append(np.var(consulta))

  i += 1

print("Média dos Tempos Médios de Consulta:", sumMediasPostgreSQL/nCons)
```

## Separa as consultas por tipo de consulta simplificada

```python
consultaSimplif = ['I', 'I', 'R', 'R', 'R', 'R', 'A', 'A', 'A', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C']

consAgroupSimplif = []
consIndicSimplif = []

i = 0
pos = -1
for cons in consultaSimplif:
  if(cons not in consAgroupSimplif):
    consAgroupSimplif.append(cons)
    consIndicSimplif.append([i])
    pos += 1
  else:
    consIndicSimplif[pos].append(i)
  i += 1


mediasTotaisMySQLSimplif = []
mediasTotaisPostgreSQLSimplif = []

somasVarMySQL = []
somasVarPostgreSQL = []

dpMySQL = []
dpPostgreSQL = []

for x in consIndicSimplif:
  mediaTotalMySQLSimplif = 0
  mediaTotalPostgreSQLSimplif = 0
  somaVarMySQL = 0
  somaVarPostgreSQL = 0

  for pos in x:
    mediaTotalMySQLSimplif += mediasMySQL[pos]
    mediaTotalPostgreSQLSimplif += mediasPostgreSQL[pos]
    somaVarMySQL += varMySQL[pos]
    somaVarPostgreSQL += varPostgreSQL[pos]

  mediaTotalMySQLSimplif /= len(x)
  mediaTotalPostgreSQLSimplif /= len(x)
  somaVarMySQL /= len(x)
  somaVarPostgreSQL /= len(x)

  dpMySQL.append(np.sqrt(somaVarMySQL))
  dpPostgreSQL.append(np.sqrt(somaVarPostgreSQL))

  mediasTotaisMySQLSimplif.append(mediaTotalMySQLSimplif)
  mediasTotaisPostgreSQLSimplif.append(mediaTotalPostgreSQLSimplif)

print("Médias (MySQL e PostgreSQL, respectivamente):")
print(mediasTotaisMySQLSimplif)
print(mediasTotaisPostgreSQLSimplif)
print()
print("Desvios Padrões (MySQL e PostgreSQL, respectivamente):")
print(dpMySQL)
print(dpPostgreSQL)
```

## Separa as consultas por tipo de consulta específica

```python
consultaN = ['I', 'I', 'R', 'R', 'R', 'R', 'A', 'A', 'A', 'CCP', 'CCP', 'CCP', 'CCP', 'CCP', 'CCP', 'CNP', 'CNP', 'CNP', 'CS', 'CS', 'CS', 'CS']

consAgroup = []
consIndic = []

i = 0
pos = -1
for cons in consultaN:
  if(cons not in consAgroup):
    consAgroup.append(cons)
    consIndic.append([i])
    pos += 1
  else:
    consIndic[pos].append(i)
  i += 1

mediasTotaisMySQL = []
mediasTotaisPostgreSQL = []

for x in consIndic:
  mediaTotalMySQL = 0
  mediaTotalPostgreSQL = 0
  for pos in x:
    mediaTotalMySQL += mediasMySQL[pos]
    mediaTotalPostgreSQL += mediasPostgreSQL[pos]

  mediaTotalMySQL /= len(x)
  mediaTotalPostgreSQL /= len(x)

  mediasTotaisMySQL.append(mediaTotalMySQL)
  mediasTotaisPostgreSQL.append(mediaTotalPostgreSQL)
```

## Barplot da comparação entre os tempos de execução baseado no tipo de operação

```python
fig = plt.figure(figsize =(9, 5.5), dpi=150)

width = 0.4

plt.bar(range(len(consAgroup)), mediasTotaisMySQL[:], width, color ='tab:orange')
plt.bar([x + width for x in range(len(consAgroup))], mediasTotaisPostgreSQL[:], width)

plt.xlabel("Tipo de Operação")
plt.ylabel("Tempo de Execução Médio (s)")
plt.xticks([n + width/2 for n in range(len(consAgroup))], [x for x in consAgroup])
plt.legend(["MySQL", "PostgreSQL"])
plt.show()
```
